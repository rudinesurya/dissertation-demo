```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import random

from scipy import fftpack
```

```python
def plotTimeSeries(t,x):
    fig, ax = plt.subplots()
    ax.plot(t, x)
    ax.set_xlabel('Time [s]')
    ax.set_ylabel('Signal amplitude')
    plt.show()
    
def plotFFT(freqs, X):
    fig, ax = plt.subplots()

    ax.plot(freqs, np.abs(X))
    ax.set_xlabel('Frequency in Hertz [Hz]')
    ax.set_ylabel('Frequency Domain (Spectrum) Magnitude')
    plt.show()
```

Test with a simple sin wave to see if the fft is working as expected

```python
f = 5  # Frequency, in cycles per second, or Hertz
f_s = 100  # Sampling rate, or number of measurements per second
t = np.linspace(0, 1, f_s)
x = np.sin(f * 2 * np.pi * t)

plotTimeSeries(t,x)
```

```python
X = fftpack.fft(x)
freqs = fftpack.fftfreq(len(x)) * f_s

plotFFT(freqs, X)
```

we should get back the original signal

```python
x = fftpack.ifft(X)

plotTimeSeries(t,x)
```

```python
df = pd.read_csv('data.csv',nrows=1000)

df.drop('Unnamed: 0', axis=1, inplace=True) #drop unused columns
target = df[['y']] #keep the target in a new dataframe
df.drop('y', axis=1, inplace=True) #drop unused columns
df.head()
```

y contains the category of the 178-dimensional input vector. 

5 - eyes open, means when they were recording the EEG signal of the brain the patient had their eyes open 

4 - eyes closed, means when they were recording the EEG signal the patient had their eyes closed 

3 - Yes they identify where the region of the tumor was in the brain and recording the EEG activity from the healthy brain area 

2 - They recorder the EEG from the area where the tumor was located 

1 - Recording of seizure activity 

```python
def analyze(i):
    print('Analyzing row:', i)
    x = df.iloc[i,:]
    f_s = 178  # Sampling rate, or number of measurements per second
    t = np.linspace(0, 1, f_s)

    plotTimeSeries(t,x)
    
    X = fftpack.fft(x)
    freqs = fftpack.fftfreq(len(x)) * f_s

    plotFFT(freqs, X)
    
    x = fftpack.ifft(X)
#     print('signal obtained using inverse FFT')
#     plotTimeSeries(t,x)
```

```python
seizures = target.loc[target['y']==1] # Recording of seizure activity
nonseizures = target.loc[target['y']!=1] # Recording of non-seizure activity
```

```python
print('Seizure data')
for item in random.sample(list(seizures.index),k=4):
    analyze(item)
```

```python
print('Non-Seizure data')
for item in random.sample(list(nonseizures.index),k=4):
    analyze(item)
```

```python
# Import the slearn utility to compare algorithms
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn import neighbors
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.decomposition import PCA

def runAllClassifiers_(Xtrain, Xtest, ytrain, ytest):
    # Prepare an array with the 2 algorithms
    models = []
    models.append(('NB', GaussianNB()))
    models.append(('SVC', SVC(gamma='auto')))
    models.append(('knn3', neighbors.KNeighborsClassifier(n_neighbors=3)))
    models.append(('knn9', neighbors.KNeighborsClassifier(n_neighbors=9)))
    models.append(('knn21', neighbors.KNeighborsClassifier(n_neighbors=21)))
    models.append(('RForest', RandomForestClassifier(n_estimators=100)))
    models.append(('AdaBoost', AdaBoostClassifier(n_estimators=100, learning_rate=0.5)))
    
    # Prepare the configuration to run the test
    seed = 42
    results = []
    names = []
    nSplits = 10
    
    # Fit the classifiers and output the crossVal results
    for name, model in models:
        # We use k-fold cross validation, where 10% of the training data is held back each time 
        kfold = model_selection.KFold(n_splits=nSplits, random_state=seed)
        crossValResults = model_selection.cross_val_score(
            model, Xtrain, np.ravel(ytrain), cv=kfold, scoring='accuracy')
        results.append(crossValResults)
        names.append(name)
        # Now print the results
        print("%s: %f (%f)" % (name, crossValResults.mean(), crossValResults.std()))
        
    
    # boxplot algorithm comparison
    fig = plt.figure()
    fig.suptitle('Algorithm Comparison')
    ax = fig.add_subplot(111)
    plt.boxplot(results)
    ax.set_xticklabels(names)
    plt.show()
```

```python
def runAllClassifiers(X, y):
    seed=42
    
    #apply pca to reduce the dimension while still retaining 99% of the features
    pca = PCA(0.99, whiten=True)
    xpca = pca.fit_transform(X)
    
    # split the data into training and test
    Xtrain, Xtest, ytrain, ytest = train_test_split(xpca, y, random_state=seed, stratify=target)
    runAllClassifiers_(Xtrain, Xtest, ytrain, ytest)
```

```python
def convertToFreq(df):
    dff = pd.DataFrame(data=None, columns = df.columns, index = df.index)
    for i in range(df.shape[0]):
        x = df.iloc[i,:]
        X = fftpack.fft(x)
        dff.loc[i] = X.real
        
    return dff
```

```python
dff = convertToFreq(df)
dff.head()
```

```python
runAllClassifiers(df, target) # classifying using time domain
```

```python
runAllClassifiers(dff, target) # classifying using freq domain
```
